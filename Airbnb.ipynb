## Chart - 13

```python
#Correlation Heatmap nsight: Relationship between price, reviews, availability, etc.
import numpy as np

# Select numerical columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Remove infinite values if any
numeric_df = numeric_df.replace([np.inf, -np.inf], np.nan)

# Compute correlation matrix
corr_matrix = numeric_df.corr()

plt.figure(figsize=(12,8))

sns.heatmap(corr_matrix,
            cmap='coolwarm',
            annot=True,
            fmt=".2f",
            linewidths=0.5)

plt.title("Correlation Heatmap: Relationship Between Numerical Variables",
          fontsize=20, fontstyle="italic", fontweight="bold", color="blue")

plt.show()
```

## 1Ô∏è‚É£ Why this chart?

A heatmap is used to visualize the strength and direction of relationships between multiple numerical variables at once.

## 2Ô∏è‚É£ Insights found?

It shows correlations between price, number of reviews, availability, minimum nights, and other numeric variables.
Strong positive or negative values indicate meaningful relationships.

## 3Ô∏è‚É£ Business impact?

Yes, it helps identify key factors affecting pricing, demand, and availability.
Weak or negative correlations may indicate inefficiencies or areas needing strategic improvement.
